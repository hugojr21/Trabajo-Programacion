```python
def crear_matriz(h, d):
    lista_principal = []
    for i in range(h):
        lista_principal.append([0] * d)
    return lista_principal

def mostrar_estado(m):
    print("\nPlan de reserva")
    print("********************************")
    for i in range(len(m)):
        print("Habitación", i + 1, m[i])
    print("********************************")

def archivos(nombre, datos):
    f = open(nombre, "w")
    f.write(datos)
    f.close()

    f = open(nombre, "r")
    print(f.read())
    f.close()

def programa():
    hotel = []
    habitaciones = 0
    dias = 0

    while True:
        print("\nSISTEMA DE PLANIFICACIÓN DE RESERVACIÓN DE HABITACIONES DE HOTEL")
        print("1. Generar plan de reservas")
        print("2. Ingresar costo de reservas por habitación y día")
        print("3. Visualizar ingreso total de una habitación")
        print("4. Visualizar ingreso total por día")
        print("5. Salir del programa")

        opcion = input("Seleccione su opción: ")

        if opcion == "1":
            habitaciones = int(input("Ingrese número de habitaciones: "))
            dias = int(input("Ingrese número de días de reserva: "))
            while dias > 5:
                print("Error, el máximo son 5 días.")
                dias = int(input("Ingrese número de días de reserva: "))

            hotel = crear_matriz(habitaciones, dias)
            mostrar_estado(hotel)

        elif opcion == "2":
            if len(hotel) == 0:
                print("Debe generar el plan primero.")
            else:
                for i in range(habitaciones):
                    print("Habitación", i + 1)
                    for j in range(dias):
                        while True:
                            costo = int(input("Ingrese costo del día " + str(j+1) + ": "))
                            if costo >= 45000 and costo <= 80000:
                                hotel[i][j] = costo
                                break
                            else:
                                print("Error, el costo debe estar entre $45.000 y $80.000.")
                mostrar_estado(hotel)

        elif opcion == "3":
            if len(hotel) == 0:
                print("No hay datos.")
            else:
                h_num = int(input("Ingrese la habitación a calcular y visualizar: "))
                if h_num >= 1 and h_num <= habitaciones:
                    total = 0
                    info = ""
                    for j in range(dias):
                        precio = hotel[h_num-1][j]
                        info = info + "Habitación " + str(h_num) + ", día " + str(j+1) + ": $" + str(precio) + "\n"
                        total = total + precio
                    info = info + "Total ingresos de la Habitación " + str(h_num) + " $: " + str(total)
                    archivos("ingreso_habitacion.txt", info)
                else:
                    print("Error, habitación no válida.")

        elif opcion == "4":
            if len(hotel) == 0:
                print("No hay datos.")
            else:
                d_num = int(input("Ingrese el día a calcular y visualizar: "))
                if d_num >= 1 and d_num <= dias:
                    total_d = 0
                    info_d = ""
                    for i in range(habitaciones):
                        precio_d = hotel[i][d_num-1]
                        info_d = info_d + "Habitación " + str(i+1) + ", día " + str(d_num) + ": $" + str(precio_d) + "\n"
                        total_d = total_d + precio_d
                    info_d = info_d + "Total ingresos del día " + str(d_num) + " $: " + str(total_d)
                    archivos("ingreso_dia.txt", info_d)
                else:
                    print("Error, día no válido.")

        elif opcion == "5":
            print("Fin de la ejecución del programa")
            break

programa()
```
